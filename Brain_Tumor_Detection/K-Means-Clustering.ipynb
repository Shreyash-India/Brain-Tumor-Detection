```python
## Brain Tumor Detection using K-Means Clustering
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread(r'C:\Users\KIIT\Tensorflow\Brain_Tumor\Brain_Tumor_Detection\brain_tumor_dataset\yes\Y255.jpg', 0)
plt.imshow(img, cmap='gray')
```

```python
height, width = img.shape
print("Height of Image (in pixels) ",height)
print("Width of Image (in pixels) ",width)
```

```python
## Pixels Range of Iamge [0,255]
np.unique(img)
```

```python
## Flatten 2-D image into 1-D
img_flatten = img.reshape(height * width, 1)
```

# K Means Clustering Model

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=4)
model.fit(img_flatten)
```

```python
## No of Clusters
clusters = model.labels_
print('No of Unique Clsuters ', np.unique(clusters))
print('No of Unique Clsuters ', np.count_nonzero(np.unique(clusters)) + 1)
```

```python
## No of pixels Belonging to each Clusters
pixels_in_each_clusters = np.bincount(clusters)
print('No of Pixels from the image belonging to each Clusters ', pixels_in_each_clusters)
```

```python
## Data Visualization
label2D = clusters.reshape(height, width)

## In vicinity of Tumor
print(img[200:210, 110:120])
print(label2D[200:210, 110:120])
```

```python
## Cluster Visualization [4 clusters]
plt.imshow(label2D)
plt.colorbar()
plt.show()
```
